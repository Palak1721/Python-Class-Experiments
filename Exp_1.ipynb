###6. Write a program to computethe length of the hypotenuse (c) of a right triangle using Pythagoras theorem.  

```python
import math

# input sides of right triangle
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))

# hyotenuse
c = math.sqrt(a**2 + b**2)

print(f"The hypotenuse is: {c:.2f}")
```

###7. Write a program to find simple interest.

```python
# Input principal, rate, and time
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time (years): "))

# Simple interest
simple_interest = (principal * rate * time) / 100

print(f"Simple Interest: ₹{simple_interest:.2f}")
print(f"Total Amount: ₹{principal + simple_interest:.2f}")
```

###8. Write a program to find area of triangle when length of sides are given.

```python
import math

# Input three sides
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))
s = (a + b + c) / 2

# Calculate area using Heron's formula
area = math.sqrt(s * (s - a) * (s - b) * (s - c))

print(f"Area of triangle: {area:.2f} sq.units")
```

###9. Write a program to convert given seconds into hours, minutes and remaining seconds.

```python
time_in_seconds = int(input("Enter seconds: "))

# Calculate hours, minutes, and remaining seconds
hours = time_in_seconds // 3600
remaining =time_in_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

print(f"{time_in_seconds} seconds = {hours}h {minutes}m {seconds}s")
```

###12. Write a program to print truth table for bitwise operators (&, | and ^ operators)

```python
print("Truth Table")
print(f"{'A':<5} {'B':<5} {'A&B':<8} {'A^B':<8} \n")
print("-"*70, "\n")

for i in [0,1]:
    for j in [0,1]:
        print(f"{i: <5} {j: <5} {i and j :<8} {i or j:<8}")
```

###13. Write a program to find left shift and right shift values of a given number.

```python
num = int(input("Enter a number: "))
shift = int(input("Enter number of positions to shift: "))

left_shift = num << shift
right_shift = num >> shift

print("Left Shift Value:", left_shift)
print("Right Shift Value:", right_shift)
```
