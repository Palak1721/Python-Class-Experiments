###4. Find the greatest among three numbers assuming no two values are same.  

```python
# Read three numbers from the user
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Check if first number is the greatest
if a > b and a > c:
    print("Greatest number is:", a)

# Check if second number is the greatest
elif b > a and b > c:
    print("Greatest number is:", b)

# If above conditions are false, third number is the greatest
else:
    print("Greatest number is:", c)
```

Compute the discriminant

Decide the nature of roots:

D > 0 â†’ two real and distinct roots

D = 0 â†’ two real and equal roots

D < 0 â†’ imaginary (complex) roots

```python
import math  # Needed for square root

# Read coefficients a, b, and c
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

# Calculate the discriminant
d = b*b - 4*a*c

# Check nature of roots
if d > 0:
    # Two real and distinct roots
    root1 = (-b + math.sqrt(d)) / (2*a)
    root2 = (-b - math.sqrt(d)) / (2*a)
    print("Roots are real and distinct")
    print("Root 1:", root1)
    print("Root 2:", root2)

elif d == 0:
    # Two real and equal roots
    root = -b / (2*a)
    print("Roots are real and equal")
    print("Root:", root)

else:
    # Imaginary roots
    real_part = -b / (2*a)
    imag_part = math.sqrt(-d) / (2*a)
    print("Roots are imaginary")
    print("Root 1:", real_part, "+", imag_part, "i")
    print("Root 2:", real_part, "-", imag_part, "i")
```

###6. Find whether a given year is a leap year or not.

```python
# Read year from the user
year = int(input("Enter a year: "))

# Check leap year conditions
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

###7. Write a program which takes any date as input and display next date of the calendar
e.g.
I/P: day=20 month=9 year=2005  
O/P: day=21 month=9 year 2005

```python
# Take input from user
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Increase day by 1
day = day + 1

# Check for February
if month == 2:
    # Leap year check
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        if day > 29:
            day = 1
            month = 3
    else:
        if day > 28:
            day = 1
            month = 3

# Months with 31 days
elif month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10:
    if day > 31:
        day = 1
        month = month + 1

# Months with 30 days
elif month == 4 or month == 6 or month == 9 or month == 11:
    if day > 30:
        day = 1
        month = month + 1

# December
elif month == 12:
    if day > 31:
        day = 1
        month = 1
        year = year + 1

# Print next date
print("Next Date:")
print("day =", day, "month =", month, "year =", year)
```

###8. Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage.
CGPA=percentage/10  
CGPA range:  
0 to 3.4 -> F  
3.5 to 5.0->C+  
5.1 to 6->B  
6.1 to 7-> B+  
7.1 to 8-> A  
8.1 to 9->A+  
9.1 to 10-> O (Outstanding)  



Sample Gradesheet  
Name: Rohit Sharma   
Roll Number: R17234512  
SAPID: 50005673   
Sem: 1    Course: B.Tech. CSE   AI&ML

Subject name: Marks   
PDS:   70  
Python:  80  
Chemistry:  90  
English:  60  
Physics:  50  


Percentage: 70%  
CGPA:7.0   
Grade:B+   

```python
# Input student details
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAP ID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

# Input marks of five subjects
pds = int(input("Enter marks of PDS: "))
python = int(input("Enter marks of Python: "))
chem = int(input("Enter marks of Chemistry: "))
eng = int(input("Enter marks of English: "))
phy = int(input("Enter marks of Physics: "))

# Calculate total marks
total = pds + python + chem + eng + phy

# Calculate percentage
percentage = total / 5

# Calculate CGPA
cgpa = percentage / 10

# Decide grade based on CGPA
if cgpa >= 0 and cgpa <= 3.4:
    grade = "F"
elif cgpa >= 3.5 and cgpa <= 5.0:
    grade = "C+"
elif cgpa >= 5.1 and cgpa <= 6.0:
    grade = "B"
elif cgpa >= 6.1 and cgpa <= 7.0:
    grade = "B+"
elif cgpa >= 7.1 and cgpa <= 8.0:
    grade = "A"
elif cgpa >= 8.1 and cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"

# Print Grade Sheet
print("\n------ GRADE SHEET ------")
print("Name:", name)
print("Roll Number:", roll, " SAPID:", sapid)
print("Sem:", sem, " Course:", course)

print("\nSubject : Marks")
print("PDS:", pds)
print("Python:", python)
print("Chemistry:", chem)
print("English:", eng)
print("Physics:", phy)

print("\n")
print("Percentage:", percentage, "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```
